# DIFFERENT TEST PERFORMED ON THE GIVEN DATA

```python
from scipy.stats import chi2_contingency



chitable = pd.crosstab(dataset.Attrition, dataset.Age)

stats1, p1, dof1, expected1 = chi2_contingency(chitable)
print(stats1, p1)


if p1 < 0.05 :
    print("Thus reject the Ho")
else:
    print("Thus Accept the Ho")
# in the result below we can say that we reject the Ho due to p < 0.05
```
